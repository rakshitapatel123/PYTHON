## COVID-19 Data Dashboard


````python
import pandas as pd
import matplotlib.pyplot as plt
````

````python
data = {
    "Date": pd.date_range(start="2020-01-01", periods=10, freq="D"),
    "Country": ["India", "India", "India", "USA", "USA", "USA", "Brazil", "Brazil", "Brazil", "Brazil"],
    "Confirmed": [10, 20, 50, 5, 15, 25, 2, 10, 20, 40],
    "Deaths": [0, 1, 2, 0, 1, 2, 0, 1, 2, 3],
    "Recovered": [0, 5, 20, 0, 7, 15, 0, 3, 10, 20]
}

df = pd.DataFrame(data)
# print(df)
df["Date"] = pd.to_datetime(df["Date"])
# print("Total Cases by Country:\n", df.groupby("Country")["Confirmed"].sum())
# print()
# print("Total Deaths by Country:\n", df.groupby("Country")["Deaths"].sum())
# print()
# print("Total Recovered by Country:\n", df.groupby("Country")["Recovered"].sum())

df["Country"].value_counts().head()
india_cases = df[df["Country"]=="India"].set_index("Date")["Confirmed"]
usa_cases = df[df["Country"]=="USA"].set_index("Date")["Confirmed"]

plt.plot(india_cases, label="India")
plt.plot(usa_cases, label="USA")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")
plt.title("COVID-19 Confirmed Cases Over Time")
plt.legend()
plt.show()

deaths = df.groupby("Country")["Deaths"].sum()
deaths.plot(kind="bar", color="red", figsize=(6,4))
plt.title("Total COVID-19 Deaths by Country")
plt.ylabel("Deaths")
plt.show()

df["Active"] = df["Confirmed"] - df["Recovered"] - df["Deaths"]
print(df.head(10))
````

## Student Exam 

````python
import numpy as np
import pandas as pd

# Create random data
np.random.seed(42)
students = ["S1", "S2", "S3", "S4", "S5"]
subjects = ["Math", "Science", "English"]

# Generate random marks using NumPy
marks = np.random.randint(50, 100, size=(len(students), len(subjects)))

# Create DataFrame
df = pd.DataFrame(marks, columns=subjects, index=students)

# Add Average column
df["Average"] = df.mean(axis=1)

print("📊 Student Marks DataFrame:\n", df)

# Find topper
topper = df.loc[df["Average"].idxmax()]
print("\n🏆 Topper:\n", topper)

````

````python
# Sample data
products = ["Laptop", "Mobile", "Tablet", "Headphones", "Charger"]
months = ["Jan", "Feb", "Mar", "Apr", "May"]

# Generate random sales using NumPy
sales = np.random.randint(100, 500, size=(len(products), len(months)))

# Create DataFrame
df = pd.DataFrame(sales, columns=months, index=products)

# Add Total Sales
df["Total"] = df.sum(axis=1)

print("📊 Sales Data:\n", df)

# Find best-selling product
best_product = df["Total"].idxmax()
print("\n🏆 Best-Selling Product:", best_product)

# Monthly total sales
print("\n📈 Monthly Sales:\n", df.sum())
````

````python
# Generate dates
dates = pd.date_range(start="2022-01-01", periods=10)

# Simulate stock prices using NumPy
np.random.seed(0)
open_prices = np.random.randint(100, 200, size=10)
close_prices = open_prices + np.random.randint(-10, 10, size=10)
volume = np.random.randint(1000, 5000, size=10)

# Create DataFrame
df = pd.DataFrame({
    "Date": dates,
    "Open": open_prices,
    "Close": close_prices,
    "Volume": volume
})

# Set Date as index
df.set_index("Date", inplace=True)

# Moving Average
df["Close_MA3"] = df["Close"].rolling(window=3).mean()

print("📊 Stock Data:\n", df)
````

## 2. Profit Margin (Profit / Sales)

````python
df["ProfitMargin"] = df["Profit"] / df["Sales"]
profit_margin_summary = df.groupby("Category")["ProfitMargin"].mean()
print("\nAverage Profit Margin by Category:\n", profit_margin_summary)
````

## 3. Top 10 Customers by Sales & Profit

````python
top_customers_sales = df.groupby("CustomerID")["Sales"].sum().sort_values(ascending=False).head(10)
top_customers_profit = df.groupby("CustomerID")["Profit"].sum().sort_values(ascending=False).head(10)
print("Top 10 Customers by Sales:\n", top_customers_sales)
print("\nTop 10 Customers by Profit:\n", top_customers_profit)
````

## 4. Monthly Sales Trend (using resample)

````python
df["OrderDate"] = pd.to_datetime(df["OrderDate"], errors="coerce")
monthly_sales = df.set_index("OrderDate").resample("M")["Sales"].sum()
print("Monthly Sales Trend:\n", monthly_sales)
import matplotlib.pyplot as plt

monthly_sales.plot(kind="bar", figsize=(10,5), title="Monthly Sales Trend")
plt.ylabel("Total Sales")
plt.show()
````

## (A) Cohort Analysis: Customer Retention by First Purchase Month

````python
df["OrderMonth"] = df["OrderDate"].dt.to_period("M")
first_purchase = df.groupby("CustomerID")["OrderMonth"].min()
df["CohortMonth"] = df["CustomerID"].map(first_purchase)
df["CohortIndex"] = (df["OrderMonth"] - df["CohortMonth"]).apply(lambda x: x.n)
cohort_data = df.groupby(["CohortMonth", "CohortIndex"])["CustomerID"].nunique().reset_index()
cohort_pivot = cohort_data.pivot(index="CohortMonth", columns="CohortIndex", values="CustomerID")
print("Cohort Analysis (Customer Retention Table):\n", cohort_pivot)
````
